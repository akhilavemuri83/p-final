```python
def classification_prediction(models,x,y):
    def cl_report(models,x,y):
        c_s={}
        x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.23,random_state=7)
        for model in models:
            m=model()
            m.fit(x_train,y_train)
            y_predict=m.predict(x_test)
            c_s[str(m)]={"model":m,
                        "traing_score"         :m.score(x_train,y_train),
                        "testing_score"        :m.score(x_test,y_test),
                        "precition_score"      :precision_score(y_test,y_predict),
                        "recall_score"         :recall_score(y_test,y_predict),
                        "confussion_matrix"    :confusion_matrix(y_test,y_predict),
                        "classification_report":classification_report(y_test,y_predict)}            
        all_models=input("do you want to see the all models performance(y/n)")            
        if all_models=="y":
            for i in c_s:
                for j,k in c_s[i].items():
                    print(j,"----->",k)
        return c_s
    def save_model(s):
        def best_m_save(s,score):
                for i,j in s[sorted(s,key=lambda x:s[x][score])[-1]].items():
                    print(i,"--------->",j)
                save=input("do you want to save this best model(y/n)")
                if save.lower()=="y":
                    md=s[sorted(s,key=lambda x:s[x][score])[-1]]['model']
                    with open("pr_model","wb") as f:         
                        pk.dump(md,f)
                        print(f"our is {s[sorted(s,key=lambda x:s[x][score])[-1]]['model']} model is saved\n ....Thank you.... ")
        while True:
            ask=input("""Do you want know the best model by
                                press 1 for "testing_score"
                                press 2 for "recall_score"
                                press 3 for "precision_score" 
                                press 4 for if don't want to print anything
                                """)        
            
            if ask=="1":
                best_m_save(s,"testing_score")
                break
            elif ask=="2":
                best_m_save(s,"recall_score")
                break
            elif ask=="3":
                best_m_save(s,"precition_score") 
                break
            elif ask=="4":
                print("Thank you .....")
                break
            else:
                print("sorry! you have pressed wrong button pleace enter correct button")
                continue
    im_data=input("""this data is imbalence data:
                          press 1 if it our data is a imbalence data
                          press 2 if it our data is not a balence data
                          """)
    if im_data=="1":
        try:
            s=cl_report(models,x,y)
            save_model(s)
        except Exception as p:
            print(p)
            print("sorry there is something worg in your code pleace check it once")

    else:
        try:
            sm=SMOTE()
            x_new,y_new=sm.fit_resample(x,y)
            s=cl_report(models,x_new,y_new)
            save_model(s)
        except Exception as p:
            print(p)
            print(" sorry! there is something wrong in your code pleace check it once")
```
